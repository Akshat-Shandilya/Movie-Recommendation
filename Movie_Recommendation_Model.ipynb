```python
import pandas as pd
import numpy as np
```

```python
movie_dataframe1 = pd.read_csv('tmdb_5000_movies.csv') 
movie_dataframe1.sample(4)
```

```python
movie_dataframe2 = pd.read_csv('tmdb_5000_credits.csv')
movie_dataframe2.sample(4)
```

```python
movie_dataframe3 = movie_dataframe1[['genres','id','keywords','overview','title']]
movie_dataframe4 = movie_dataframe2[['title','cast','crew']]
```

```python
movie_dataframe = movie_dataframe3.merge(movie_dataframe4)
movie_dataframe.sample(4)
```

```python
movie_dataframe.isnull().sum()
```

```python
movie_dataframe.dropna( inplace=True ) 
movie_dataframe.isnull().sum()
```

```python
movie_dataframe
```

```python
movie_dataframe.duplicated().sum()
```

```python
movie_dataframe.iloc[0].genres
```

```python
import ast
ast.literal_eval(movie_dataframe.iloc[0].genres)
```

```python
def extract( obj ) :
    list = []
    for i in ast.literal_eval( obj ) :
        list.append( i['name'] )
    return list
```

```python
movie_dataframe['genres'] = movie_dataframe['genres'].apply(extract)
movie_dataframe.sample(4)
```

```python
movie_dataframe['keywords'] = movie_dataframe['keywords'].apply(extract)
movie_dataframe.sample(4)
```

```python
def extract_actors ( obj ) :
    list = []
    number_of_actors = 0
    for i in ast.literal_eval( obj ) :
        if number_of_actors != 5 :
            list.append( i['name'] )
            number_of_actors += 1
    return list
```

```python
movie_dataframe['cast'] = movie_dataframe['cast'].apply(extract_actors)
movie_dataframe.sample(4)
```

```python
def extract_crew ( obj ) :
    list = []
    for i in ast.literal_eval( obj ) :
        if i['job'] == 'Writer' or 'Director' or 'Screenplay' :
            if i['name'] not in list :
                list.append( i['name'] )
    return list
```

```python
movie_dataframe['crew'] = movie_dataframe['crew'].apply(extract_crew)
movie_dataframe.sample(4)
```

```python
movie_dataframe['overview'] = movie_dataframe['overview'].apply( lambda x : x.split() )
movie_dataframe.sample(4)
```

```python
movie_dataframe['genres'] = movie_dataframe['genres'].apply( lambda x : [ i.replace(' ' , '_') for i in x ] )
movie_dataframe['cast'] = movie_dataframe['cast'].apply( lambda x : [ i.replace(' ' , '_') for i in x ] )
movie_dataframe['crew'] = movie_dataframe['crew'].apply( lambda x : [ i.replace(' ' , '_') for i in x ] )
movie_dataframe['keywords'] = movie_dataframe['keywords'].apply( lambda x : [ i.replace(' ' , '_') for i in x ] )
movie_dataframe.sample(4)
```

```python
movie_dataframe['tags'] = movie_dataframe['genres'] + movie_dataframe['keywords'] + movie_dataframe['crew'] + movie_dataframe['cast']
movie_dataframe.sample(4)
```

```python
new_movie_dataframe = movie_dataframe[['title','tags','id']]
new_movie_dataframe.sample(4)
```

```python
new_movie_dataframe['tags'] = new_movie_dataframe['tags'].apply( lambda x : " ".join(x) )
new_movie_dataframe.sample(4)
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem_text ( text ) :
    list = []
    for i in text.split() :
        list.append(ps.stem(i))
    return list
```

```python
new_movie_dataframe['tags'] = new_movie_dataframe['tags'].apply( stem_text )
new_movie_dataframe.sample(4)
```

```python
new_movie_dataframe['tags'] = new_movie_dataframe['tags'].apply( lambda x : " ".join(x) )
new_movie_dataframe.sample(4)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer( max_features = 5000 , stop_words = 'english' )
```

```python
tags_vectorization = cv.fit_transform( new_movie_dataframe['tags'] ).toarray()
tags_vectorization
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity_array = cosine_similarity( tags_vectorization )
similarity_array.shape
```

```python
def recommend(movie):
    index = new_movie_dataframe[new_movie_dataframe['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity_array[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new_movie_dataframe.iloc[i[0]].title)
```

```python
recommend('The Shawshank Redemption')
```
